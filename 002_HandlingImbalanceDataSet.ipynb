```python
import pandas as pd
```

```python
df = pd.read_csv('/home/jangark20/Jupyter/002_test/creditcard.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
## dependent feature is class 
df['Class'].value_counts()
```

0 has huge records and 1 has less records, so imbalanced dataset

```python
## Independent and Depending features
x = df.drop('Class',axis=1)
y = df.Class
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
from sklearn.model_selection import KFold,GridSearchCV 
import numpy as np
```

```python
log_class = LogisticRegression()
grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}
cv=KFold(n_splits=5,random_state=None,shuffle=False)
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7)
```

```python
clf = GridSearchCV(log_class,grid,cv=cv,n_jobs=-1,scoring='f1_macro')
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier 
classifier = RandomForestClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
## Over Sampling 
y_train.value_counts()
```

```python
class_weight=dict({0:1,1:543})
```

```python
from sklearn.ensemble import RandomForestClassifier 
classifier = RandomForestClassifier(class_weight=class_weight)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from collections import Counter
Counter(y_train)
```

```python
from collections import Counter
from imblearn.under_sampling import NearMiss
ns = NearMiss(sampling_strategy='auto') 
x_train_ns, y_train_ns = ns.fit_resample(x_train, y_train)
print(f'The no. of classes befor fit {Counter(y_train)}')
print(f'The no. of classes after fit {Counter(y_train_ns)}')
```

```python
from sklearn.ensemble import RandomForestClassifier 
classifier = RandomForestClassifier()
classifier.fit(x_train_ns,y_train_ns)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

### Over sampling 

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(sampling_strategy='auto')
x_train_os , y_train_os = os.fit_resample(x_train,y_train)
print(f'The no. of classes before fit {Counter(y_train)}')
print(f'The no of classes after fit {Counter(y_train_os)}')
```

```python
from sklearn.ensemble import RandomForestClassifier 
classifier = RandomForestClassifier()
classifier.fit(x_train_os,y_train_os)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

### SMOTETomek

```python
from imblearn.combine import SMOTETomek
```

```python
sm = SMOTETomek(sampling_strategy='auto')
x_train_sm , y_train_sm = sm.fit_resample(x_train,y_train)
print(f'The no. of classes before fit {Counter(y_train)}')
print(f'The no of classes after fit {Counter(y_train_sm)}')
```

```python
from sklearn.ensemble import RandomForestClassifier 
classifier = RandomForestClassifier()
classifier.fit(x_train_sm,y_train_sm)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
